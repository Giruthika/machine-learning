```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd 
df=pd.read_csv("Social_Network_Ads.csv")
df
```

```python
df=df.drop(columns=["User ID"])
```

```python
df
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
independent=df[["Age", "EstimatedSalary", "Gender_Male"]]
independent
```

```python
dependent=df[["Purchased"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=1/3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
print("Confusion Matrix:\n",cm)
```

```python
clf_report=classification_report(y_test,y_pred)
print("Classification Report:\n",clf_report)
```

```python
import joblib
joblib.dump(classifier,'random_forest_model.pk1')
print("Model saved s random_forest_model.pk1")
```

```python
loaded_model=joblib.load('random_forest_model.pk1')
print("Model loaded successfully.")
```

```python
y_pred_loaded=loaded_model.predict(x_test)
print("Predictions from loaded model:\n",y_pred_loaded)
```
